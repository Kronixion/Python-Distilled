```python
#1. Primitives, Variables and Expressions

#1.3 Primitives (are the basic data types of Python)
integer: int = 42
real_number: float = 4.2
text: str = "Hello World!"
boolean: bool = True

print(f"Integer: {integer}")
print(f"Float  : {real_number}")
print(f"String : {text}")
print(f"Boolean: {boolean} \n")

#Simple Interest algorithm
def interest(principal: int, rate: float, num_years: int) -> None: 
    year: int = 1
    while year <= num_years:
        principal = principal * (1 + rate)
        print(f"Year: {year} - Principal: {principal:0.2f}")
        year += 1

print("Calling interest algorithm with: principle = 1000, rate = 0.05 and the number of years = 5")
interest(1000, 0.05, 5)
```

```python
#1.5 Conditionals and Control flow

#1.5.1 if statement

a = 5
b = 6

print(f"a: {a} - b: {b}\n")
print("Is a higher than b?")
if a > b:
    print("Computer says yes.\n")
else:
    print("Computer says no.\n")

#1.5.2 elif statement
print("Elif statement\n")
if a > b:
    print("a is higher than b\n")
elif a < b:
    print("a is lower than b\n")
else:
    print("a is equal to b\n")

#1.5.3 Ternary operator (in-line if)
print("\nTernary operator (in-line if)\n")
maxval = a if a > b else b
print(f"maxval: {maxval}\n")

#1.5.4 While statement
while (a < 10):
    a += 1
    print(f"a: {a}")

#1.5.5 Walrus operator (:=)
print("\nWalrus operator (:=)\n")
a = 5
while (a := a+1) < 10:
    print(f"a: {a}")

#1.5.6 Break statement (break)
print("\nBreak statement\n")
a = 5
while a < 10:
    a += 1
    if a == 7:
        break
    print(f"a: {a}")

#1.5.7 Continue statement
print("\nContinue statement\n")
a = 5
while a < 10:
    a += 1
    if a == 7:
        continue
    print(f"a: {a}")
```

```python
#1.7 File Input and Output

#1.7.1 File extraction program example:
with open('misc_files/data.txt') as file: # <- I love context managers, they open and close files without you needing to do so.
    for line in file:
        print(line, end='') # <- end='' omits the extra new line.

#The open() returns a new file object which is assigned to the file object.

#Same example without the context manager:
print("\n")
file = open('misc_files/data.txt') # <- Here we have to open the file, but once we're done with it we'll have to close it ourselves
for line in file:
    print(line, end='')
file.close() # <- Because we don't rely on the context manager we have to close the file such that it doesn't remain open, because if another resource tries to open the file it won't be able to do so.

#As a rule of thumb, it is better to use the with statement (context manager), due to the fact that you might forget to close programs, files, connections etc.

#1.7.2 Reading file in chunks
#To read a file in chunks the read() method must be used with a size parameter. Mainly used for large files. 
#A good question that you can think about is: What does the size parameter represent? Take a look at the example presented below.

print('\n')
with open('misc_files/data.txt') as file:
    while (chunk := file.read(2)):
        print(chunk, end='\n')

#When the end of a file is reached, read() returns an empty string
#An alternative example:

with open('misc_files/data.txt') as file:
    while True:
        chunk = file.read(2)
        if not chunk:
            break
        print(chunk, end='\n')

#Up to this point we've practiced reading from files (the in part), now we'll go over to writing information in files (the out part).
year = 1
numyears = 5
principal = 1000
rate = 0.05

with open('misc_files/out.txt', 'wt') as out: # <- Do you know what the 'wt' mode does? What is the default modeand what are some other file modes?
    while year <= numyears:
        principal = principal * (1 + rate)
        print(f'{year:>3d} {principal:0.2f}', file=out)
        year += 1

#Additionally file objects support a write method. The print example from above can be written like this:
#out.write('{year:>3d} {principal:0.2f}')

#1.7.3 Writing information from the console
# If you've ever wandered, how can I make a program to run from the console?
# The answer is, you can use the input() method to type data interactively in the console. (Similar to the Python REPL - Read Evaluate Print Loop)

name = input("Enter your name: ")
print(f"Hello {name}")
```

```python
#1.8 Lists
#Lists are an ordered collection of arbitrary objects.
#A simple list representation is a grocery list, where you have arbitrary objects such as vegetables, dairy products, spices etc.

#1.8.1 Creating a list
#A list is created in Python using [] (square brackets)
grocery_list = ["Minced Beef", "Carrots", "Chilli", "Salt"]

#1.8.2 Accessing and replacing an item from the list
#To access an item from a list, an integer index can be used.
#Similar to strings the index of the first element in the list starts from 0.
print(f"The item from the grocery list at index 0 is: {grocery_list[0]}")

#To replace a field in the list, you can use the assign operator (=) while accessing the element at a given index.
grocery_list[0] = "Chicken Breast" # <- We don't want to eat red meat and we've switched it with white meat as we want to be healthier.
print(f"The replaced item from the grocery list at index 0 is {grocery_list[0]}")

#1,8.3 Appending an item to a list
#If we want to buy more ingredients and add them to the end of our grocery_list we can use the append(item) method
grocery_list.append("Apple Scented Water")
print(f"We've added the following item at the end of our list: {grocery_list[-1]}")

#1.8.4 Iterating over a list
#Similar to iterating over characters in a string, we can also iterate over the items in our list.
print("\nGrocery List: ")
for item in grocery_list:
    print(f" - {item}")

#1.8.5 Slicing
#Similar to strings, we can use slicing to read a portion of a list or to reassign that portion.
print("\n")
print(f"The first three items from the grocery_list are: {grocery_list[0:2]}")
previous_grocery_items = grocery_list.copy() # <- Copies the list. (Explanation will come in later chapters)
grocery_list[0:3] = ["Apple", "Cucumber", "Milk"]
print(f"We've replaced the first the first three items from: {previous_grocery_items} with {grocery_list[0:3]} resulting in {grocery_list}")

#1.8.6 Concatenation
#To concatenate lists we can use + operator.
print("\n")
second_grocery_list = ["Kiwi", "Blood Oranges", "Monster"]
grocery_list = grocery_list + second_grocery_list # <- Now we don't want to have multiple grocery lists lying around, let's combine them into one.
print(f"The combined grocery list is: {grocery_list}")

#1.8.7 Creating lists
#Now creating an empty list can be done in the following two methods
names = []
names = list()

#list is actually a class and a constructor for creating an empty list data structure (It is mainly a data type, because it is an object)
#Keep in mind that the difference between a data type and a data structure conceptually is, data types describe what something is, while data structures describe how something is ordered.

#Here are several additional examples of creating lists
letters = list('Calin')
print(f"Name letters: {letters}")

random_list = [1, 'Dave', 3.14, ['Mark', 7, 9, [100, 101]], 10]
print(f"Random List: {random_list}")

#The good thing about lists is that you can also represent matrices.
matrix = [[1,2],[3,4]]
print(f"Matrix: {matrix}")

#Let's say that we want to view the list as two columns and two rows and we want to compute their product.
#How can we do that?
#The answer is easy, we use an expression within the list while looping through the matrix.

column_product = [row[0] * row[1] for row in matrix]
print(f"Column product (1st code implementation): {column_product}")

#A similar code representation is

column_product = []
for row in matrix:
    column_product.append(row[0] * row[1])

#This method takes 3 lines of code, while the previous one took one line of code.
print(f"Column product (2nd code implementation): {column_product}")

```
