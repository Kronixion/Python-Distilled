```python
#1. Primitives, Variables and Expressions

#1.3 Primitives (are the basic data types of Python)
integer: int = 42
real_number: float = 4.2
text: str = "Hello World!"
boolean: bool = True

print(f"Integer: {integer}")
print(f"Float  : {real_number}")
print(f"String : {text}")
print(f"Boolean: {boolean} \n")

#Simple Interest algorithm
def interest(principal: int, rate: float, num_years: int) -> None: 
    year: int = 1
    while year <= num_years:
        principal = principal * (1 + rate)
        print(f"Year: {year} - Principal: {principal:0.2f}")
        year += 1

print("Calling interest algorithm with: principle = 1000, rate = 0.05 and the number of years = 5")
interest(1000, 0.05, 5)
```

```python
#1.5 Conditionals and Control flow

#1.5.1 if statement

a = 5
b = 6

print(f"a: {a} - b: {b}\n")
print("Is a higher than b?")
if a > b:
    print("Computer says yes.\n")
else:
    print("Computer says no.\n")

#1.5.2 elif statement
print("Elif statement\n")
if a > b:
    print("a is higher than b\n")
elif a < b:
    print("a is lower than b\n")
else:
    print("a is equal to b\n")

#1.5.3 Ternary operator (in-line if)
print("\nTernary operator (in-line if)\n")
maxval = a if a > b else b
print(f"maxval: {maxval}\n")

#1.5.4 While statement
while (a < 10):
    a += 1
    print(f"a: {a}")

#1.5.5 Walrus operator (:=)
print("\nWalrus operator (:=)\n")
a = 5
while (a := a+1) < 10:
    print(f"a: {a}")

#1.5.6 Break statement (break)
print("\nBreak statement\n")
a = 5
while a < 10:
    a += 1
    if a == 7:
        break
    print(f"a: {a}")

#1.5.7 Continue statement
print("\nContinue statement\n")
a = 5
while a < 10:
    a += 1
    if a == 7:
        continue
    print(f"a: {a}")
```

```python
#1.7 File Input and Output

#1.7.1 File extraction program example:
with open('misc_files/data.txt') as file: # <- I love context managers, they open and close files without you needing to do so.
    for line in file:
        print(line, end='') # <- end='' omits the extra new line.

#The open() returns a new file object which is assigned to the file object.

#Same example without the context manager:
print("\n")
file = open('misc_files/data.txt') # <- Here we have to open the file, but once we're done with it we'll have to close it ourselves
for line in file:
    print(line, end='')
file.close() # <- Because we don't rely on the context manager we have to close the file such that it doesn't remain open, because if another resource tries to open the file it won't be able to do so.

#As a rule of thumb, it is better to use the with statement (context manager), due to the fact that you might forget to close programs, files, connections etc.

#1.7.2 Reading file in chunks
#To read a file in chunks the read() method must be used with a size parameter. Mainly used for large files. 
#A good question that you can think about is: What does the size parameter represent? Take a look at the example presented below.

print('\n')
with open('misc_files/data.txt') as file:
    while (chunk := file.read(2)):
        print(chunk, end='\n')

#When the end of a file is reached, read() returns an empty string
#An alternative example:

with open('misc_files/data.txt') as file:
    while True:
        chunk = file.read(2)
        if not chunk:
            break
        print(chunk, end='\n')

#Up to this point we've practiced reading from files (the in part), now we'll go over to writing information in files (the out part).
year = 1
numyears = 5
principal = 1000
rate = 0.05

with open('misc_files/out.txt', 'wt') as out: # <- Do you know what the 'wt' mode does? What is the default modeand what are some other file modes?
    while year <= numyears:
        principal = principal * (1 + rate)
        print(f'{year:>3d} {principal:0.2f}', file=out)
        year += 1

#Additionally file objects support a write method. The print example from above can be written like this:
#out.write('{year:>3d} {principal:0.2f}')

#1.7.3 Writing information from the console
# If you've ever wandered, how can I make a program to run from the console?
# The answer is, you can use the input() method to type data interactively in the console. (Similar to the Python REPL - Read Evaluate Print Loop)

name = input("Enter your name: ")
print(f"Hello {name}")
```

```python
#1.9 Tuples

#A tuple is an immutable object in Python, which packs a collection of values or objects.
#1.9.1 Creating tuples

domain = ("www","google","com") # <- This tuple represents the Subdomain, Root domain (Domain name), and Top level domain of google.
address = ("192.168.0.100", "443") # <- This is an IPV4 address and a porn number, which is mainly used for the Hyper Text Transfer Protocol Secure (HTTPS) protocol.

print(f"Domain: {domain}")
print(f"Address: {address}")

#In order to be complete the creation, we can also have 0 and 1 element tuples.
zero_element_tuple = ()
one_element_tuple = (1,)

#1.9.2 Value Retrieval
#In order to retrieve values from a tuple, an index can be used similar to a list.
#Starting element is 0
print("\n")
subdomain = domain[0]

print(f"Subdomain: {subdomain}")

#Besides the index method, which is rarer than unpacking a tuple into several variables.
print("\n")

subdomain, domain_name, top_level_domain = domain
print(f"Google domain: https://{subdomain}.{domain_name}.{top_level_domain}")

#Due to the tuples properties of immutability, values cannot be added or removed from it.
#It would be best to view tuples as an object that consists of several parts.

#1.9.3 Example (Tuples and Lists)
print("\n")
portfolio_file_path = "misc_files/portfolio.csv"

portfolio = []
with open(portfolio_file_path) as file:
    for line in file:
        row = line.split(',')
        name, shares, price = row[0], int(row[1]), float(row[2])
        holding = (name, shares, price)
        portfolio.append(holding)

#The portofolio list is created which looks like a two-dimmentional array.
#A nice analogy would be an excel table of n rows and 3 columns.
print(f"Portfolio: {portfolio}")

#1.9.4 For loop unpacking.
#Now that we every item in the portfolio is a list we can unpack it as variables directly in the for loop.
print("\n")

total = 0.0
for name, shares, price in portfolio:
    total += shares * price

#Or we can use a in-list for loop.
total = sum([shares * price for _, shares, price in portfolio]) # <- The variable _ indicates a discarded value.

print(f"Total portfolio value: {total}")
```

```python
#1.10 Sets
#A set is an unordered collection of unique objects.
#We mainly use the to find distinct values or to managed problems related to membership.

#1.10.1 Creating a set
#There are two methods: {items} or set()
#!Do not confuse the curly brakets with a items in with a dictionary and if you only type {} and assign it to a variable, it will create an empty dictionary.
#!For empty sets use set()

names1 = {"IBM", "MSFT", "AA"}
names2 = set(["IBM", "MSFT", "HPE", "IBM", "CAT"])

print(f"Names1: {names1}")
print(f"Names2: {names2}") # <- Because "IBM" appears twice, we can see that only one is used, because a set contains only unique objects.

#1.10.2 Set comprehension
#For this example, please run the previous code cell.
names = { s[0] for s in portfolio } # <- here we can observe that the names of the portfolio have been turned into a set, meaning that any name duplicates are removed.
print(f"\nPortfolio names: {names}")

#1.10.3 Set operations.
#As in mathematics the set has several common operations.
print("\n")
print(f"Union - t | s: {names1 | names2}") # Groups the two sets together, while keeping the unique items.
print(f"Intersection - t & s: {names1 & names2}") # Keeps the items that are the same from the two sets.
print(f"Difference - t - s: {names1 - names2}") # Removes the items from set1 that are found in set2.
print(f"Symmetric difference - t ^ s: {names1 ^ names2}") # Removes the items from both sets that are the same and keeps the ones that aren't.

#Besides the set operations, we have some common ones, such as add, update, remove, and discard.
example_set = set()

example_set.add("DIS") # Adds a single item to the set.
example_set.update({"JJ", "GE", "ACME"}) # Adds multiple items to a set.
print(f"\nAdding values to example_set: {example_set}")
example_set.remove("DIS") # Removes DIS or it raises a KeyValue error if absent.
example_set.discard("JJ") # Removes JJ if it exists.
print(f"\nRemoving values from example_set: {example_set}")
```

```python
#1.11 Dictionaries
#A dictionary is a mapping between key and values. 
#A good analogy is the lexical dictionary book as it maps words to definitions.
#The word is the key, while the definition is the value.

#1.11.1 Creating dictionaries
#To create a dictionary you can either have {} or dict().

empty_dict = {}
s = {
    'name': 'GOOG',
    'shares': 100,
    'price': 590.12
}
print(f"Manually defined dictionary: {s}")

#1.11.2 Accessing values
name = s['name']
cost = s['shares'] * s['price']
print(f"\nname: {name} - cost: {cost}")

#You can also use the get() method.
print("\n")
print(f"Get method with key and default parameter: {s.get('name', "ASM")}")
print(f"Get method with key and no default parameter: {s.get('rate')}") # If you don't provide a parameter it will return None

#1.11.3 Removing key value pairs from a dictionary
print("\n")
del s['name']
print(f"Dictionary: {s}") # It can be observed that the key 'name' along with it's value is being removed.

#del s['rate'] <- will return a KeyError as 'rate' does not exist.

#As a nice piece of information in dictionaries keys can contain immutables objects, 
#but mutable data structures such as lists, sets, and dictionaries can't be used as keys.

#Dictionaries are often used as building blocks for various algorithms and data-handling problems.
#One such problem is tabulation. One such example, is, how you could count the total number of shares for each stock.

portfolio = [
    ('ACME', 50, 92.34),
    ('IBM', 75, 102.25),
    ('PHP', 40, 74.50),
    ('IBM', 50, 124.75)
]

#Dictionary comprehension
total_shares = { s[0]: 0 for s in portfolio } # Retrieving all the keys and placing them in a dictionary with a default value of 0

for name, shares, _ in portfolio:
    total_shares[name] += shares

print(f"\nTotal shares: {total_shares}")

#A dict() method can also be used with multiple parameters or multiple variants of a parameter.
#For example, we can pass it a list of tuples, which are transformed into a dictionary of keys and values.

pairs = [('IBM', 125), ('ACME', 50), ('PHP', 40)]
dictionary = dict(pairs)

print(f"\nDictionary: {dictionary}")

#1.11.4 Retrieving all the keys from a dictionary.
#There are two possible methods in which this can be done.
#The first method is to type cast a list on a dictionary. For example:

keys1 = list(dictionary)
print(f"\nList of keys: {keys1}")

#The second method is to call the .keys() method on the dictionary which returns a key view.

keys2 = dictionary.keys()
print(f"\nKeys view: {keys2}")

#The difference between the first method and the second one is that the key view is tied to the dictionary, meaning that changes are reflected.

dictionary["GOOG"] = 130
print(f"\nList of keys contains: {keys1}") # It will not contain GOOG as a key
print(f"\nKey view now contains: {keys2}") # It will also contains GOOG as a key

#Obtaining values or key value pairs.
#To obtain the values of a dictionary, use the .values() method.

print(f"\nThe values of a dictionary: {dictionary.values()}") # Returns a view of dict_values.

#To obtain the key-value pairs of a dictionary, use the .items() method.

print(f"\nThe items of a dictionary: {dictionary.items()}") # Returns a view of dict_items.
```
