```python
#1. Primitives, Variables and Expressions

#1.3 Primitives (are the basic data types of Python)
integer: int = 42
real_number: float = 4.2
text: str = "Hello World!"
boolean: bool = True

print(f"Integer: {integer}")
print(f"Float  : {real_number}")
print(f"String : {text}")
print(f"Boolean: {boolean} \n")

#Simple Interest algorithm
def interest(principal: int, rate: float, num_years: int) -> None: 
    year: int = 1
    while year <= num_years:
        principal = principal * (1 + rate)
        print(f"Year: {year} - Principal: {principal:0.2f}")
        year += 1

print("Calling interest algorithm with: principle = 1000, rate = 0.05 and the number of years = 5")
interest(1000, 0.05, 5)
```

```python
#1.4 Python Operators

#1.4.1 Arithmetic Operators

first_number = 7
second_number = 4

print(f"Addition            - x + y: {first_number + second_number}")
print(f"Substration         - x - y: {first_number - second_number}")
print(f"Multiplication      - x * y: {first_number * second_number}")
print(f"Division            - x / y: {first_number / second_number}")
print(f"Truncating division - x // y: {first_number // second_number}")
print(f"Power               - x ** y: {first_number ** second_number}")
print(f"Modulo              - x % y: {first_number % second_number}")
print(f"Unary minus         - -x: {-first_number}")
print(f"Unary plus          - +x: {+first_number}")

#1.4.2 Numerical Operations (not Python Operators)
print("\n")
print(f"Absolute value - abs(x): {abs(first_number)}")
print(f"Div Mod - divmoc(x,y): {divmod(first_number, second_number)}")
print(f"(x**y) % modulo - pow(x,y [,modulo]): {pow(first_number, second_number)}")
print(f"Rounding to the nearest multiple of 10 to the n-th power - round(x,[n]) {round(first_number/second_number,2)}")

#1.4.3 Bitwise Operators
print("\n")

x = 0b11000000
mask = 0b10000001
number_of_positions = 3

print(f"x   : {bin(x)}")
print(f"mask: {bin(mask)}")
print(f"number_of_positions: {bin(number_of_positions)}")

print("\n")

print(f"Left Shift (Shifts the bits to the left by a number of positions)   - x << y: {bin(x << number_of_positions)}")
print(f"Right Shift (Shifts the bits to the right by a number of positions) - x >> y: {bin(x >> number_of_positions)}")
print(f"Bitwise AND      - x & y: {bin(x & mask)}")
print(f"Bitwise OR       - x | y: {bin(x | mask)}")
print(f"Bitwise XOR      - x ^ y: {bin(x ^ mask)}")
print(f"Bitwise Negation - ~x: {bin(~x)}")

#1.4.4 Comparison Operators
print("\n")

six = 6
seven = 7

print(f"Equal to                 - x == y: {six == seven}")
print(f"Not Equal to             - x != y: {six != seven}")
print(f"Less than                - x < y : {six < seven}")
print(f"Greater than             - x > y : {six > seven}")
print(f"Greater than or equal to - x >= y: {six >= seven}")
print(f"Lesser than or equal to  - x <= y: {six <= seven}")

#1.4.5 Logical Operators
print("\n")

x = True
y = False

print(f"x or y : {x or y}")
print(f"x and y: {x and y}")
print(f"not x  : {not x}")
```

```python
#1.5 Conditionals and Control flow

#1.5.1 if statement

a = 5
b = 6

print(f"a: {a} - b: {b}\n")
print("Is a higher than b?")
if a > b:
    print("Computer says yes.\n")
else:
    print("Computer says no.\n")

#1.5.2 elif statement
print("Elif statement\n")
if a > b:
    print("a is higher than b\n")
elif a < b:
    print("a is lower than b\n")
else:
    print("a is equal to b\n")

#1.5.3 Ternary operator (in-line if)
print("\nTernary operator (in-line if)\n")
maxval = a if a > b else b
print(f"maxval: {maxval}\n")

#1.5.4 While statement
while (a < 10):
    a += 1
    print(f"a: {a}")

#1.5.5 Walrus operator (:=)
print("\nWalrus operator (:=)\n")
a = 5
while (a := a+1) < 10:
    print(f"a: {a}")

#1.5.6 Break statement (break)
print("\nBreak statement\n")
a = 5
while a < 10:
    a += 1
    if a == 7:
        break
    print(f"a: {a}")

#1.5.7 Continue statement
print("\nContinue statement\n")
a = 5
while a < 10:
    a += 1
    if a == 7:
        continue
    print(f"a: {a}")
```
